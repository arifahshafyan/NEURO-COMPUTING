```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, mean_squared_error as mse
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('data/iris.csv')
data = dataset.iloc[:, 0:-1]  
label = dataset.iloc[:, -1]  

data_train, data_test, label_train, label_test = train_test_split(data, label, test_size=0.2, random_state=42)
iterations = np.array([45,75, 100,125, 150, 200, 250, 300,350])
errors = []
train_accuracies = []
test_accuracies = []

print('======< IRIS Dataset >======')
for i in range(len(iterations)):
    model = MLPClassifier(hidden_layer_sizes=(10), max_iter=iterations[i],
                          activation='logistic', learning_rate_init=0.1, random_state=42)
    model.fit(data_train, label_train)
    pred_train = model.predict(data_train)
    error_train = mse(label_train, pred_train)
    errors.append(error_train)
    pred_test = model.predict(data_test)
    accuracy_train = accuracy_score(label_train, pred_train)
    accuracy_test = accuracy_score(label_test, pred_test)
    train_accuracies.append(accuracy_train)
    test_accuracies.append(accuracy_test)
    print(f'Iteration {iterations[i]}: Train Accuracy = {accuracy_train:.2f}, '
          f'Test Accuracy = {accuracy_test:.2f}, MSE = {error_train:.2f}')

plt.plot(iterations, train_accuracies, marker='o', label='Train Accuracy', color='blue')
plt.plot(iterations, test_accuracies, marker='o', label='Test Accuracy', color='green')
plt.xlabel('Number of Iterations')
plt.ylabel('Accuracy')
plt.title('Training and Test Accuracy on IRIS')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score, mean_squared_error

dataset_iris = pd.read_csv('data/iris.csv')
data_iris = dataset_iris.iloc[:, 0:-1]
label_iris = dataset_iris.iloc[:, -1]

data_iris_train, data_iris_test, label_iris_train, label_iris_test = train_test_split(data_iris, label_iris, test_size=0.3, random_state=42)
scaler = StandardScaler()
data_iris_train = scaler.fit_transform(data_iris_train)
data_iris_test = scaler.transform(data_iris_test)

# Train MLP model
model_iris = MLP(hidden_layer_sizes=(20), activation='logistic', learning_rate_init=0.05, max_iter=500, random_state=42, warm_start=True)
intervals = [10, 20, 30, 40, 50, 100, 150, 200, 280, 300, 400, 501]

for i in intervals:
    model_iris.max_iter = i
    model_iris.fit(data_iris_train, label_iris_train)
    pred_iris = model_iris.predict(data_iris_test)
    accuracy_iris = accuracy_score(label_iris_test, pred_iris)
    mse_iris = mean_squared_error(label_iris_test, pred_iris)
    print(f'Iteration {i}: Accuracy = {accuracy_iris:.2f}, MSE = {mse_iris:.2f}')

# Plot training loss at each iteration
train_losses = [model_iris.loss_ for i in range(1, 501)]  # Capture the training loss for plotting
plt.figure(figsize=(10, 6))
plt.plot(range(1, 501), train_losses, marker='o', linestyle='-', color='b')
plt.title('Training Loss vs. Iterations')
plt.xlabel('Number of Iterations')
plt.ylabel('Training Loss')
plt.grid(True)
plt.show()
```
